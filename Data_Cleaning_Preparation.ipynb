```python
#import libraries 
import pandas as pd
```

```python
# Load the original and extracted datasets
original_data = pd.read_csv("NY-House-Dataset.csv")  # Replace with your file path
extracted_data = pd.read_csv("housing_geocode_extraction.csv")  # Replace with your API-extracted dataset
```

```python
# Merge datasets on LATITUDE and LONGITUDE
merged_data = original_data.merge(extracted_data, on=['LATITUDE', 'LONGITUDE'], suffixes=('_original', '_extracted'))
```

```python
# Columns to validate
columns_to_compare = ['STATE', 'LOCALITY', 'SUBLOCALITY', 'POSTCODE']

# Compare the columns and flag discrepancies
for col in columns_to_compare:
    merged_data[f'{col}_match'] = merged_data[f'{col}_original'] == merged_data[f'{col}_extracted']

# Calculate accuracy for each column
accuracy_results = {
    col: merged_data[f'{col}_match'].mean() * 100 for col in columns_to_compare
}

# Print accuracy results
print("Accuracy Results:")
for col, acc in accuracy_results.items():
    print(f"{col}: {acc:.2f}%")
```

```python
# Save discrepancies to a file for review
discrepancies = merged_data[~merged_data[[f'{col}_match' for col in columns_to_compare]].all(axis=1)]
discrepancies.to_csv("discrepancies.csv", index=False)
print("\nDiscrepancies saved to 'discrepancies.csv'")
```
